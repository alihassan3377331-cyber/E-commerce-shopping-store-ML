### Data Preprocessing: Cleaning the 'price' column

The `price` column is currently a string and contains non-numeric characters such as 'Tk', commas, and units like '/month'. To use it for analysis or model training, we need to convert it to a numerical format.

## convert string into integer

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

mart['location'] = mart['location'].astype(str)
mart['location'] = le.fit_transform(mart['location']).astype(int)
mart['category'] = mart['category'].astype(str)
mart['category'] = le.fit_transform(mart['category']).astype(int)
mart['slug'] = mart['slug'].astype(str)
mart['slug'] = le.fit_transform(mart['slug']).astype(int)
mart['isMember'] = mart['isMember'].astype(str) # Convert to string first
mart['isMember'] = le.fit_transform(mart['isMember']).astype(int)
mart['description'] = mart['description'].astype(str)
mart['description'] = le.fit_transform(mart['description']).astype(int)
mart['imgUrl'] = mart['imgUrl'].astype(str)
mart['imgUrl'] = le.fit_transform(mart['imgUrl']).astype(int)
mart['timeStamp']=mart['timeStamp'].astype(str)
mart['timeStamp']=le.fit_transform(mart['timeStamp']).astype(int)
mart['isLocalJob']=mart['isLocalJob'].astype(str)
mart['isLocalJob']=le.fit_transform(mart['isLocalJob']).astype(int)
mart['isFeaturedMember']=mart['isFeaturedMember'].astype(str)
mart['isFeaturedMember']=le.fit_transform(mart['isFeaturedMember']).astype(int)
mart['isAuthDealer']=mart['isAuthDealer'].astype(str)
mart['isAuthDealer']=le.fit_transform(mart['isAuthDealer']).astype(int)

# Correctly clean and convert 'price' column to a numerical type (float)
mart['price'] = mart['price'].astype(str).str.replace('Tk', '', regex=False)
mart['price'] = mart['price'].str.replace(',', '', regex=False)
mart['price'] = mart['price'].str.replace(' /month', '', regex=False) # Handle '/month' suffix
mart['price'] = pd.to_numeric(mart['price'], errors='coerce') # Convert to numeric, coerce non-convertibles to NaN
mart['price'] = mart['price'].fillna(0).astype(float) # Fill NaNs and ensure float type
mart['price'] = np.log1p(mart['price']) # Apply log1p transformation to handle skewness and zeros

mart['adType']=mart['adType'].astype(str)
mart['adType']=le.fit_transform(mart['adType']).astype(int)
mart['isVerified']=mart['isVerified'].astype(str)
mart['isVerified']=le.fit_transform(mart['isVerified']).astype(int)
mart['isJobAd']=mart['isJobAd'].astype(str)
mart['isJobAd']=le.fit_transform(mart['isJobAd']).astype(int)
mart['isUrgentAd']=mart['isUrgentAd'].astype(str)
mart['isUrgentAd']=le.fit_transform(mart['isUrgentAd']).astype(int)
mart['isTopAd']=mart['isTopAd'].astype(str)
mart['isTopAd']=le.fit_transform(mart['isTopAd']).astype(int)
mart['details']=mart['details'].astype(str)
mart['details']=le.fit_transform(mart['details']).astype(int)
mart['shopName']=mart['shopName'].astype(str)
mart['shopName']=le.fit_transform(mart['shopName']).astype(int)
mart['isDoorstepDelivery']=mart['isDoorstepDelivery'].astype(str)
mart['isDoorstepDelivery']=le.fit_transform(mart['isDoorstepDelivery']).astype(int)
mart['isDeliveryFree']=mart['isDeliveryFree'].astype(str)
mart['isDeliveryFree']=le.fit_transform(mart['isDeliveryFree']).astype(int)
mart['membershipLevel']=mart['membershipLevel'].astype(str)
mart['membershipLevel']=le.fit_transform(mart['membershipLevel']).astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

# Ensure all object columns in x_train are numerical before scaling
# Create a new LabelEncoder instance to avoid conflicts from previous fittings if any
le_x_train = LabelEncoder()

for col in x_train.select_dtypes(include=['object']).columns:
    x_train[col] = le_x_train.fit_transform(x_train[col].astype(str)).astype(int)

for col in x_test.select_dtypes(include=['object']).columns:
    # Use the same fitted LabelEncoder from x_train to transform x_test
    # Handle potential new categories in test set if they appear by fitting again or using a different strategy if needed
    # For this case, refitting on x_test column is a simple approach assuming new categories are rare or can be handled.
    # A more robust approach for production would be to fit on entire dataset or use handle_unknown='ignore' if available.
    x_test[col] = le_x_train.fit_transform(x_test[col].astype(str)).astype(int)

# Fit the scaler on the training data and transform both training and test data
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

## Hyperparameter Tuning with GridSearchCV

Hyperparameter tuning is the process of selecting the optimal set of hyperparameters for a machine learning model. Instead of manually guessing, we can use techniques like `GridSearchCV` to systematically search for the best parameters.

`GridSearchCV` exhaustively searches through a specified subset of hyperparameters, training and evaluating a model for each combination. This helps in finding the hyperparameters that yield the best performance metrics (e.g., R-squared).

### KNN Regressor Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for KNN
knn_param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Initialize GridSearchCV for KNN
knn_grid_search = GridSearchCV(knn, knn_param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=1)

# Fit GridSearchCV to the scaled training data
knn_grid_search.fit(x_train_scaled, y_train)

# Print the best parameters and best score for KNN
print("Best parameters for KNN:", knn_grid_search.best_params_)
print("Best R-squared score for KNN (on validation sets):", knn_grid_search.best_score_)

# Get the best KNN model
best_knn = knn_grid_search.best_estimator_

# Make predictions with the best KNN model
y_pred_knn_tuned = best_knn.predict(x_test_scaled)

# Evaluate the best KNN model on the test set
r2_knn_tuned = r2_score(y_test, y_pred_knn_tuned)
print("R-squared score for tuned KNN on test set:", r2_knn_tuned)
```

### Random Forest Regressor Hyperparameter Tuning
